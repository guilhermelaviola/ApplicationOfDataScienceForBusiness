# **Time Series Analysis**
Time series analysis is a statistical and machine learning approach that helps companies understand past behavior and predict future outcomes by analyzing historical data over time. It involves collecting and organizing relevant data, visualizing it to identify trends, seasonality, and outliers, and applying decomposition techniques to separate the series into trend, seasonal, and residual components. Understanding these elements enables more accurate forecasting for applications such as demand forecasting, inventory management, and financial planning. Tools like Python and libraries such as Pandas, NumPy, Matplotlib, and Statsmodels support efficient analysis and modeling, allowing organizations to make informed, data-driven decisions, optimize operations, and adapt to market changes.

```python
# Importing all the necessary libraries and resources:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## **Example: Time series Analysis**
Time series analysis is a powerful tool for understanding and predicting business behavior. By breaking down time series into trends, seasonality, and residuals, we gain valuable insights that help with strategic decision-making. Using tools like Python and libraries like statsmodels, we can perform these analyses effectively and accurately.

By understanding the following components, we can apply time series analysis in practical ways to business. For example:
1. **Sales Planning**: Identifying growth trends helps with strategic sales planning and resource allocation.
2. **Inventory Management**: Seasonality analysis allows you to adjust inventory levels based on seasonal peaks and valleys.
3. **Error Reduction**: By considering residuals, it is possible to minimize forecast errors and improve the accuracy of business strategies.



```python
# Generating dummy data: a 24-month time series with sales ranging from 80,000 to 120,000 units:
dates = pd.date_range(start='2023-01-01', periods=24, freq='M')
sales = np.random.randint(80, 120, size=(24,))

# Creating the DataFrame with the data above:
data = pd.DataFrame({'Date': dates, 'Sales': sales})
```

```python
# Initial setup and plotting:
plt.figure(figsize=(12,6))
plt.plot(data['Date'], data['Sales'], label='Monthly Sales')
plt.title('Monthly Sales Time Series')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Decomposing the time series:
result = seasonal_decompose(data['Sales'], model='multiplicative', period=12)

# Plotting the components:
result.plot()
plt.show()
```

```python
# Applying a moving average to smooth the time series:
data['Smoothed_Sales'] = data['Sales'].rolling(window=3).mean()

# Plotting the original and smoothed time series:
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Sales'], label='Monthly Sales')
plt.plot(data['Date'], data['Smoothed_Sales'], label='Smoothed Sales', linestyle='--')
plt.title('Monthly Sales Time Series with Smoothing')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Decomposing the time series:
result = seasonal_decompose(data['Sales'], model='multiplicative', period=12)

# Plotting the components:
result.plot()
plt.show()
```

```python
# Subplots and Visualization:
plt.figure(figsize=(12, 8))

# Trend:
plt.subplot(3, 1, 1)
plt.plot(result.trend, label='Trend', color='blue')
plt.title('Trend')

# Seasonality:
plt.subplot(3, 1, 2)
plt.plot(result.seasonal, label='Seasonality', color='green')
plt.title('Seasonality')

# Residuals:
plt.subplot(3, 1, 3)
plt.plot(result.resid, label='Residuals', color='red')
plt.title('Residuals')

plt.tight_layout()
plt.show()
```
